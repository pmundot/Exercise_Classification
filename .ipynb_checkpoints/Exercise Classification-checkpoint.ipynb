# Vector Equation to find angles

Instead of using the standardized points that MediaPipe Produces, I choose to apply the standardization of the points to the actual file size to get the actual image points. From there I can get actual joint angles by using three points on the data.

```python
def angle_2p_3d(a, b, c):       

    v1 = np.array([ a[0] - b[0], a[1] - b[1], a[2] - b[2]])
    v2 = np.array([ c[0] - b[0], c[1] - b[1], c[2] - b[2]])

    v1mag = np.sqrt([ v1[0] * v1[0] + v1[1] * v1[1] + v1[2] * v1[2]])
    v1norm = np.array([ v1[0] / v1mag, v1[1] / v1mag, v1[2] / v1mag])

    v2mag = np.sqrt(v2[0] * v2[0] + v2[1] * v2[1] + v2[2] * v2[2])
    v2norm = np.array([ v2[0] / v2mag, v2[1] / v2mag, v2[2] / v2mag])
    res = v1norm[0] * v2norm[0] + v1norm[1] * v2norm[1] + v1norm[2] * v2norm[2]
    angle_rad = np.arccos(res)

    return round(math.degrees(angle_rad),2)
```

# Pose Collections

When creating image classifications points, angle usage was determined. The right side of the body was also chosen for 2 reasons. One of them being that for most exercisesâ€™ pictures on the internet, the right side of the body is the most prominent and easiest to detect and two to speed up the calculation of the angles from 20 to 10. When done, the angles can then be displayed on the image for the benefits of seeing them.

```python
#Creating the list for the right side of the body
re = [12,14,16]
rs = [14,12,24]
rh = [12,24,26]
rk = [24,26,28]
ra = [26,28,32]
right_side = [re,rs,rh,rk,ra]

#Column names for future dataframes
name = ['Exercise','right_elbow','right_shoulder','right_hip','right_knee','right_ankle']

def angles_to_image(lst,img):
    one = lst[0] 
    two = lst[1]
    three = lst[2]
    cv2.circle(img,(x[one],y[one]),4,(255,0,0),cv2.FILLED)
    cv2.circle(img,(x[two],y[two]),4,(255,0,0),cv2.FILLED)
    cv2.circle(img,(x[three],y[three]),4,(255,0,0),cv2.FILLED)
    once = angle_2p_3d((x[one],y[one],z[one]),(x[two],y[two],z[two]),(x[three],y[three],z[three]))
    cv2.putText(img=img, 
                text=str(once), 
                org=(x[two], y[two]), 
                fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                fontScale=.5, color=(0, 255, 0),thickness=1)

def angles_to_df(lst,img):
    one = lst[0] 
    two = lst[1]
    three = lst[2]
    once = angle_2p_3d((x[one],y[one],z[one]),(x[two],y[two],z[two]),(x[three],y[three],z[three]))
    return(once)
```

From there we can extract the joint positions orlandmarks and calculate the given angles of each body part. How do we know which landmarks are which though? Luckly mediapipe has a guide for where each point lines up on the body. When given the land marks, we can extract them with a simple for loop and get our data points.

![pose](./images/pose_tracking_full_body_landmarks.png)

```python
image = cv2.imread('images/jump.jpg') 

RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

results = pose.process(RGB) #Data points results
ids = []
x = []
y = []
z = []
for i,poses in enumerate(results.pose_landmarks.landmark):
    h,w,d = image.shape
    x.append(int(poses.x*w))
    y.append(int(poses.y*h))
    z.append(int(poses.z*d))
    ids.append(i)
finall = np.reshape([angles_to_df(t,image) for t in right_side],(1,5))

#mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
for i in right_side:
    angles_to_image(i,image)

cv2.imwrite('images/Angels.jpeg',image)

display.Image('images/Angels.jpeg')
```

## Model Building
Now the analysis part of the data begins. The data is split into a testing set of 30% and a training set of 70%. This is to make sure we get the best amount of options possible and have a well trained data

```python
X = df_train[df_train.columns[1:]]
y = df_train[df_train.columns[0]]
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7)

print(X_train.shape), print(y_train.shape)
print(X_test.shape), print(y_test.shape)
```

Because the dataset is approximately 450 data points, k folds is also needed in the analysis to make sure overfitting doesnâ€™t occur. We can also determine who has the best accuracy with a smaller data set as well as the full dataset.

```python
def kFolds_Models(model,k=5,X=X,y=y):
    kf = KFold(n_splits=k, random_state=None)
    acc_score = []
 
    for train_index , test_index in kf.split(X):
        X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
        y_train , y_test = y[train_index] , y[test_index]

        model.fit(X_train,y_train)
        pred_values = model.predict(X_test)

        acc = metrics.accuracy_score(pred_values , y_test)
        acc_score.append(acc)

    avg_acc_score = sum(acc_score)/k

    print('accuracy of each fold - {}'.format(acc_score))
    print('Avg accuracy : {}'.format(avg_acc_score))
```

### Random Forest Classifier

```python
clf = RandomForestClassifier()
kFolds_Models(clf)
clf.fit(X_train,y_train)
ypredclf = clf.predict(X_test)
y_trainclf = clf.predict(X_train)

con = metrics.confusion_matrix(y_test, ypredclf)
print('Confusion Matrix')
print(con)
print("Test Accuracy Score %f | Train Accuracy Score %f"%(metrics.accuracy_score(y_test, ypredclf),metrics.accuracy_score(y_train,y_trainclf)))
print("Test Precision Score %f | Train Precision Score %f"%(metrics.precision_score(y_test, ypredclf,average='macro'),metrics.precision_score(y_train, y_trainclf,average='macro')))
print("Test Recall Score %f | Train Recall Score %f"%(metrics.recall_score(y_test, ypredclf,average='macro'),metrics.recall_score(y_train, y_trainclf,average='macro')))
print("Test F1 Score %f | Train F1 Score %f"%(metrics.f1_score(y_test, ypredclf,average='macro'),metrics.f1_score(y_train, y_trainclf,average='macro')))
```

### Naive Bayes Classifier

```python
nb = GaussianNB()
kFolds_Models(nb)
nb.fit(X_train,y_train)
yprednb = nb.predict(X_test)
y_trainnb = nb.predict(X_train)

con = metrics.confusion_matrix(y_test, yprednb)
print('Confusion Matrix')
print(con)
print("Test Accuracy Score %f | Train Accuracy Score %f"%(metrics.accuracy_score(y_test, yprednb),metrics.accuracy_score(y_train,y_trainnb)))
print("Test Precision Score %f | Train Precision Score %f"%(metrics.precision_score(y_test, yprednb,average='macro'),metrics.precision_score(y_train, y_trainnb,average='macro')))
print("Test Recall Score %f | Train Recall Score %f"%(metrics.recall_score(y_test, yprednb,average='macro'),metrics.recall_score(y_train, y_trainnb,average='macro')))
print("Test F1 Score %f | Train F1 Score %f"%(metrics.f1_score(y_test, yprednb,average='macro'),metrics.f1_score(y_train, y_trainnb,average='macro')))
```

### Logistical Regression

```python
lr = LogisticRegression()
kFolds_Models(lr)
lr.fit(X_train,y_train)
ypredlr = lr.predict(X_test)
y_trainlr = lr.predict(X_train)

con = metrics.confusion_matrix(y_test, ypredlr)
print('Confusion Matrix')
print(con)
print("Test Accuracy Score %f | Train Accuracy Score %f"%(metrics.accuracy_score(y_test, ypredlr),metrics.accuracy_score(y_train,y_trainlr)))
print("Test Precision Score %f | Train Precision Score %f"%(metrics.precision_score(y_test, ypredlr,average='macro'),metrics.precision_score(y_train, y_trainlr,average='macro')))
print("Test Recall Score %f | Train Recall Score %f"%(metrics.recall_score(y_test, ypredlr,average='macro'),metrics.recall_score(y_train, y_trainlr,average='macro')))
print("Test F1 Score %f | Train F1 Score %f"%(metrics.f1_score(y_test, ypredlr,average='macro'),metrics.f1_score(y_train, y_trainlr,average='macro')))
```

From the given data, Random Forest Classifier seems to perform the best amoungst them all. From here on out this is what will be used for the picture presentations and video applications.

### Conclusion
While this Classifiers was able to appropriately label each exercise it could correctly, there are still problems that need to be worked through. One of these problems is with the push-ups classification. MediaPipe has a hard time find the shoulders of a person in a horizontal position. The data will place the shoulder on the neck or middle back area skewing the angles for the shoulder. Another issue that was unforeseen was how squats and push ups had similar correlations in the arms. If This were to be done again squats would be replaced with another exercise, or more arm variation in the dataset would be used for squats. Overall, The project was a success as the model used was able to classify workouts in both picture and video to a great accuracy.
